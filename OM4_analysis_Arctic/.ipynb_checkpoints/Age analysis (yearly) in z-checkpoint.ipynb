```python
%reset
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import xarray as xr
import cartopy.crs as ccrs
import netCDF4 as nc
import matplotlib.ticker as mticker
import warnings; warnings.filterwarnings('ignore')
import cmocean
from gsw import alpha, SA_from_SP, p_from_z, CT_from_pt, beta, sigma1 
```

```python
%cd /archive/ogrp/CMIP6/OMIP/warsaw_201803_mom6_2018.04.06/OM4p25_IAF_BLING_CFC_csf/gfdl.ncrc4-intel16-prod/pp/ocean_annual_z/
fs=xr.open_dataset('ocean_annual_z.static.nc')
xh=fs['xh']
yh=fs['yh']
lon=fs['geolon']; lat=fs['geolat']
lonu=fs['geolon_u']; latu=fs['geolat_u']
lonv=fs['geolon_v']; latv=fs['geolat_v']
deptho=fs['deptho'][:,:]
print(fs)
%cd av/annual_5yr/
```

```python
data=xr.open_dataset('ocean_annual_z.2003-2007.ann.nc')['time']
i=0 
a=str(data[i].values); title=a
print(title)

fhist=nc.Dataset('ocean_annual_z.2003-2007.ann.nc')
z_l=fhist.variables['z_l'][:]
so=fhist.variables['so'][0,:,:,:]
agessc=fhist.variables['agessc'][0,:,:,:]
print(fhist)
```

```python
agessc.shape
```

```python
answer = np.empty((1080,1440)); answer[:] = np.NaN
for i in range(0,1080):
    for j in range(0,1440):
        df = pd.Series(agessc[:,i,j])
        final_index = df.last_valid_index()
        if final_index!=None:
            answer[i,j]=agessc[final_index,i,j]
            
    print(i)
```

```python
fig = plt.figure(figsize=(10, 10))
ax = plt.axes(projection=ccrs.Orthographic(0, 90))
ax.set_extent([-180, 180, 70, 90],ccrs.PlateCarree())

plt.pcolormesh(lon, lat,np.ones((1080,1440)),transform=ccrs.PlateCarree(),cmap='gray',vmin=0,vmax=2);

plt.pcolormesh(lon, lat,answer[:,:]/np.nanmax(answer),transform=ccrs.PlateCarree(),cmap=cmocean.cm.matter,vmin=0,vmax=1);

cbar=plt.colorbar(fraction=0.046, pad=0.04); cbar.ax.tick_params(labelsize=25);
gl=ax.gridlines(crs=ccrs.PlateCarree())
plt.title('Normalized age at bottom',fontsize=25)
gl.ylocator = mticker.FixedLocator([70, 75, 80, 85, 90])

ax.scatter(180, 90, marker='o',c='black', transform=ccrs.PlateCarree())
pointx=[0,0,0,0]; pointy=[70,75,80,85]; text=['70$^\circ$','75$^\circ$','80$^\circ$','85$^\circ$']
for i in range(len(pointx)):
    ax.text(pointx[i],pointy[i],text[i],ha='center', transform=ccrs.PlateCarree(),fontsize=16,color='black')
```
